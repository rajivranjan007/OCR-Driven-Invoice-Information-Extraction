```python
# Specify the filename of the uploaded image
image_path = list(uploaded.keys())[0]
```

```python
# Initialize an empty list to store OCR results
ocr_results = []
```

```python
# Define image preprocessing functions
def preprocess_image(image_path):
    # Load the image using OpenCV
    image = cv2.imread(image_path)

    # Convert the image to grayscale
    gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

    # Apply thresholding to create a binary image
    _, binary_image = cv2.threshold(gray_image, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)

    # Apply noise reduction and remove artifacts
    denoised_image = cv2.fastNlMeansDenoising(binary_image, None, h=5)

    return denoised_image
```

```python
# Process each uploaded image
for filename in uploaded.keys():
    image_path = filename

    # Preprocess the image
    preprocessed_image = preprocess_image(image_path)

    # Perform OCR
    text = pytesseract.image_to_string(Image.fromarray(preprocessed_image))
    ocr_results.append({'Filename': filename, 'Text': text})
```

```python
# Create a DataFrame from the OCR results
df = pd.DataFrame(ocr_results)
```

```python
def clean_text(text):
    # Remove special characters except spaces
    cleaned_text = re.sub(r'[^\w\s]', '', text)
    # Convert to lowercase
    cleaned_text = cleaned_text.lower()
    # Remove extra spaces
    cleaned_text = ' '.join(cleaned_text.split())
    return cleaned_text

# Apply the cleaning function to the 'Text' column
df['Cleaned Text'] = df['Text'].apply(clean_text)
```

```python
import re

def extract_invoice_date(text):
    # Regular expression pattern to match various date formats
    pattern = r'\d{4}-\d{2}-\d{2}|\d{2}/\d{2}/\d{4}|[A-Za-z]{3}\s\d{1,2},\s\d{4}|\d{1,2}\s[A-Za-z]{3},\s\d{4}|\d{1,2}-[A-Za-z]{3}-\d{4}|\d{2}-\d{2}-\d{4}'
    match = re.search(pattern, text, re.IGNORECASE)
    if match:
        return match.group(0)  # Use group 0 to get the entire matched date
    else:
        return None
```

```python
df['Invoice Date'] = df['Text'].apply(extract_invoice_date)
```

```python
def extract_gst_numbers(text):
    pattern = r'\b\d{2}[A-Za-z]{5}\d{4}[A-Za-z]{1}\d{1}[A-Za-z]{1}\d{1}\b'
    pattern1 = r'\b\d{2}[A-Za-z]{2}[A-Za-z0-9]{11}\b'

    gst_numbers = re.findall(pattern, text)
    if gst_numbers:
        return ', '.join(gst_numbers)
    else:
        gst_numbers = re.findall(pattern1, text)
        if gst_numbers:
            return ', '.join(gst_numbers)
        else:
            return None
```

```python
# Apply the function to create a new column
df['GST Num'] = df['Text'].apply(extract_gst_numbers)
```

```python
# Regex pattern for basic email validation
email_pattern = r'[A-Za-z0-9._%+-]+@[A-Za-z0-9.-]+\.[A-Za-z]{2,7}'
```

```python
# Function to extract email addresses from text
def extract_email(text):
    # Replace misplaced characters with the correct ones
    text_corrected = re.sub(r'[,]', '.', text)
    emails = re.findall(email_pattern, text_corrected)
    return ', '.join(emails) if emails else None
```

```python
# Apply the function to create a new column 'Vendor Email'
df['Vendor Email'] = df['Text'].apply(extract_email)
```

```python
# Regular expression pattern for finding 6-digit postal codes
pin_code_pattern = r'\b\d{6}\b'
```

```python
# Function to extract 6-digit postal code from text
def extract_pin_code(text):
    pin_codes = re.findall(pin_code_pattern, text)
    return pin_codes[0] if pin_codes else None
```

```python
# Regular expression pattern for Indian mobile numbers
mobile_number_pattern = r'(\+91\s?\d{10}|\d{11}|\d{10})'
```

```python
# Function to extract mobile numbers from text
def extract_mobile_numbers(text):
    mobile_numbers = re.findall(mobile_number_pattern, text)
    return ', '.join(mobile_numbers) if mobile_numbers else None
```

```python
# Apply the function to create a new column 'Mobile Numbers'
df['Mobile Numbers'] = df['Text'].apply(extract_mobile_numbers)
```

```python
# Regular expression pattern for Buyer's Order No
order_no_pattern = r"Buyer's Order No.\s*,\s*(\w+)"
```

```python
# Function to extract Buyer's Order No
def extract_buyer_order_no(text):
    match = re.search(order_no_pattern, text)
    if match:
        return match.group(1)
    else:
        return None
```

```python
# Apply the function to create a new column 'Buyer Order No'
df['Buyer Order N'] = df['Text'].apply(extract_buyer_order_no)
```

```python
import spacy

# Load the English language model
nlp = spacy.load("en_core_web_sm")
```

```python
def extract_order_numbers(text):
    doc = nlp(text)
    order_numbers = []

    for ent in doc.ents:
        if "Order" in ent.text and "No." in ent.text:
            order_numbers.append(ent.text)

    return order_numbers if order_numbers else None
```

```python
df['Order Numbers'] = df['Text'].apply(extract_order_numbers)
```
